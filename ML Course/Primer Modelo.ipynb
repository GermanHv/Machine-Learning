Modelo de regresión lineal de grados Fahrenheit a  grados Celsius

Carga de datos

```python
import pandas as pd
datos = pd.read_csv("Celsius.csv")
```

```python
datos.info()
```

```python
datos.head()
```

Visualización de los datos

```python
import seaborn as sb 

sb.scatterplot(x="Celsius", y="Fahrenheit", data=datos,
               hue="Fahrenheit", palette="coolwarm")
```

Preparación de datos

```python
# Características (X), etiqueta (y)
X = datos["Celsius"]
y = datos["Fahrenheit"]
```

```python
X
```

```python
y
```

```python
#type(X)
#type(X.values)
#X.values

#[-40, -10,   0,   8,  15, ...]
#Requiero una transformación de un arreglo con arreglos de cada uno, esto para tener varias características en modelos futuros
#[[-40], [-10],   [0],   [8],  [15], [...]]
X_procesada = X.values.reshape(-1,1)
y_procesada = y.values.reshape(-1,1)
#Esto basicamente nos da que cada valor cuenta con su propio arreglo de una sola posición y esto es requerido por modelos más complejos
```

```python
#Importar modelo de regresión lineal de sklearn
from sklearn.linear_model import LinearRegression

#Creación del modelo
modelo = LinearRegression()
```

```python
#Entrenamiento del modelo con los datos
modelo.fit(X_procesada, y_procesada)
```

```python
celsius = -53
#Predicción del modelo
prediccion = modelo.predict([[celsius]])
print(f"{celsius} grados celsius son {prediccion} grados fahrenheit")
```

```python
#evaluación del modelo 0-1  (1 es un modelo perfecto)
modelo.score(X_procesada,y_procesada)
```
